```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

```

```python
import pandas as pd
data=pd.read_csv("drive/MyDrive/news-filter-data.csv")
```

```python
import ast
res=[]
for i in data['content']:
  i=ast.literal_eval(i)
  res.append(i)
```

```python
import torch 
torch.device('cuda:0')
```

```python
res=[" ".join(i) for i in res]
```

```python
encodings=model.encode(res)
```

```python
import numpy as np
np.save("drive/MyDrive/encodings",encodings)
```
