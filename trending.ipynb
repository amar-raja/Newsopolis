```python
import pandas as pd
import pickle as p
from tqdm.notebook import tqdm
import spacy
import nltk
from nltk.corpus import stopwords
import re
import string
```

```python
df = pd.read_csv('news-full-data.csv',parse_dates=['date'])
f=open('full-content-list','rb')
l=p.load(f)
```

```python
df['unpreprocessed_headline'] = [l[i].split('\n')[0] for i in range(len(l))]
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
named_tags = []
for i in tqdm(df.unpreprocessed_headline):
    doc = nlp(i)
    named_tags.append([chunk.text for chunk in doc.noun_chunks])
```

```python
sw=stopwords.words('english')
```

```python
translate_table = dict((ord(char), None) for char in string.punctuation)
```

```python
removed=[]
for i in named_tags:
    curr=[]
    for j in range(len(i)):
        regex_sub = i[j].translate(translate_table)
        if regex_sub=='':
            continue
        if regex_sub.lower() in sw:
            continue
        curr.append(regex_sub)
    removed.append(curr)
```

```python
df['named_tags']=removed
```

```python
grouped=df.groupby(by='date')
```

```python
d=dict()
for i in grouped:
    curr_df=i[1]
    s=set()
    for j in curr_df.named_tags:
        if len(j)>=2:
            for l in range(len(j)-1):
                curr_bigram=tuple(sorted([j[l].lower(),j[l+1].lower()]))
                s.add(curr_bigram)
    for j in s:
        if j in d:
            d[j]+=1
        else:
            d[j]=1
```
